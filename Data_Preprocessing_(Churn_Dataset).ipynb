```python
import pandas as pd
#I concatinated the train and test datasets to work with the whole data
df_train = pd.read_csv('/content/churn-bigml-80.csv')
df_test = pd.read_csv('/content/churn-bigml-20.csv')

df = pd.concat([df_train, df_test], ignore_index=True)

print(df.head())
print(df.info())
```

```python
# filling numeric with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Filling categorical with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# One-hot encode (use pd.get_dummies for simplicity)
df = pd.get_dummies(df, columns=['State', 'International plan', 'Voice mail plan'], drop_first=True)
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# for a DataFrame back:
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.to_csv('X_train.csv', index=False)
```
